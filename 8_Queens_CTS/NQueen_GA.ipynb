```python
# Fitness Function
def fitness_func(chess_board):

    # Horizontal and Vertical Queen Attach Check and its accuracy
    tot_len = abs(len(chess_board))
    uniq_len = len(np.unique(chess_board))              
    row_col_clashes = tot_len - uniq_len
    clash_perct = uniq_len/tot_len
    
#     print(tot_len, uniq_len)
#     print(clash_perct)

    # Diagonal Queen Attach Check and its accuracy
    counter = 0
    for i in range(0, tot_len):
        for j in range(0, tot_len):
            if(i!=j):
                index_diff = abs(j-i)
                queen_occupation = abs(chess_board[i]-chess_board[j])
                if (queen_occupation == index_diff):
                    counter +=1
#                     print(i,j, "  ---> Clashed")
    diag_acc = (1-counter/56)
#     print(1-counter/56)
# Fitness Accuracy calculation
    consolid_acc = (diag_acc + clash_perct)/2
#     print(consolid_acc)

    return consolid_acc
```

```python
# Crossover function
def crossover(parent1, parent2):
#     2,3,4,5,6,7,8,6
#     6,0,3,4,1,9,3,1
    rand_slice = secure_random.choice(chromosome)
    child1 = parent1[0:rand_slice] + parent2[rand_slice:] 
    child2 = parent2[0:rand_slice] + parent1[rand_slice:]
    return child1, child2
```
